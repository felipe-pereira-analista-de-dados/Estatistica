```python
import pandas as pd
import requests
```

```python
response = requests.get('https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$format=json')
information = response.json()
df = pd.DataFrame(information ['value'])

pd.set_option('display.float_format', '{:.2f}'.format)


display(df)
```

```python
df.info()
```

```python
def calcular_estatisticas(coluna):
    moda = int(coluna.mode()[0])
    mediana = int(coluna.median())
    media = int(coluna.mean())
    variancia = int(coluna.var())
    desvPadrao = int(coluna.std())
    return moda, mediana, media, variancia, desvPadrao


moda_quantidade, mediana_quantidade, media_quantidade, variancia_quantidade, desvPadrao_quantidade = calcular_estatisticas(df['Quantidade'])
moda_valor, mediana_valor, media_valor, variancia_valor, desvPadrao_valor = calcular_estatisticas(df['Valor'])


print(f"Moda (Quantidade): {moda_quantidade}")
print(f"Mediana (Quantidade): {mediana_quantidade}")
print(f"Média (Quantidade): {media_quantidade}")
print(f"Variancia (Quantidade): {variancia_quantidade}")
print(f"Desvio Padrao (Quantidade): {desvPadrao_quantidade}")

print(f"Moda (Valor): {moda_valor}")
print(f"Mediana (Valor): {mediana_valor}")
print(f"Média (Valor): {media_valor}")
print(f"Variancia (Valor): {variancia_valor}")
print(f"Desvio Padrao (Valor): {desvPadrao_valor}")
```
